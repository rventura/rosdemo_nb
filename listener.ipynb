```python
import rclpy
from rclpy.node import Node
from std_msgs.msg import String
```

```python
# Initializes ROS2 library
rclpy.init()
```

```python
# Defines a callback, that is, a function that will be called upon reception of a message from topic
def callback(data):
    # Prints the string, in the data field of the message
    print(data.data)
```

```python
def listener():

    # Initialized node with name talker
    node = Node("listener")
    
    # Creates a subscriber for topic xpto, type String, and a QoS history depth of 10
    sub = node.create_subscription(String, "xpto", callback, 10)
    sub # prevent unused variable warning

    # Keep running the node, idle, while processing messages from subscribed topics
    rclpy.spin(node)
```

```python
listener()
```
