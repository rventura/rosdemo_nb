# Simple demo of a ROS subscriber

based on this ROS tutorial: https://wiki.ros.org/ROS/Tutorials/WritingPublisherSubscriber%28python%29

29-Abr-2025, Rodrigo Ventura

```python
import rospy
from std_msgs.msg import String
```

```python
# Defines a callback, that is, a function that will be called upon reception of a message from topic
def callback(data):
    # Prints the string, in the data field of the message
    print(data.data)
```

```python
def listener():
    # Initializes node with name listener
    rospy.init_node('listener', anonymous=True)

    # Subsctibes to topic called xpto, accepting type string, with callback function
    rospy.Subscriber("xpto", String, callback)

    # Keep running the node, idle, while processing messages from subscribed topics
    rospy.spin()
```

```python
listener()
```
