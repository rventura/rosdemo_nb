# Simple demo of a ROS publisher

```python
import rclpy
from rclpy.node import Node
from std_msgs.msg import String
```

```python
# Initializes ROS2 library
rclpy.init()
```

```python
def talker():

    # Initialized node with name talker
    node = Node("talker")
    
    # Creates a publisher for topic xpto, type String, and a QoS history depth of 10
    pub = node.create_publisher(String, "xpto", 10)

    # Creates a callback that sends a message
    n = 1
    def callback():
        nonlocal n
        msg = String()
        msg.data = f"Hello world #{n}"
        pub.publish(msg)
        print(f"Sent message '{msg.data}'")
        n += 1

    # Creates a timer with a constant rate of 1 Hz
    timer = node.create_timer(1, callback)

    # Main loop; exits if the node is killed through ROS
    rclpy.spin(node)
```

```python
talker()
```
