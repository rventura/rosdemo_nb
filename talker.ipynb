# Simple demo of a ROS publisher

based on this ROS tutorial: https://wiki.ros.org/ROS/Tutorials/WritingPublisherSubscriber%28python%29

29-Abr-2025, Rodrigo Ventura

```python
import rospy
from std_msgs.msg import String
```

```python
def talker():
    # Creates the publisher for topic xpto, type String
    pub = rospy.Publisher('xpto', String, queue_size=10)

    # Initialized node with name talker
    rospy.init_node('talker', anonymous=True)

    # Facility to provide constant rate of 1 Hz
    rate = rospy.Rate(1)

    # Main loop; exits if the node is killed through ROS
    while not rospy.is_shutdown():
        
        # Get the current ROS time (might be different than real time)
        t = rospy.get_time()
        
        # Publishes a simple string message
        pub.publish(f"hello world at {t}")

        # Sleeps the remaining time such that the loop iterates at 1 Hz
        rate.sleep()
```

```python
talker()
```
